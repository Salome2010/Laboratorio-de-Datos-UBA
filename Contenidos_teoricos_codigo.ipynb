 **Array Broadcasting**:

En el siguiente ej  B tiene una sola dimensión, pero tiene el mismo número de elementos que las columnas de A.
NumPy realiza un broadcasting para que B sea tratado como si tuviera la forma (2, 3), repitiendo sus valores en cada fila de A
En este caso, B es "extendido" o "broadcasted" a una forma de (2, 3) para coincidir con la forma de A, permitiendo realizar la operación de suma.

```python
A = np.array([[1, 2, 3], [4, 5, 6]]) #(2,3)
```

***Definicion media y std***

* Cuando un conjunto de datos empieza a ser más grande es conveniente definir algunos observables para resumirlos.
      1. La tendencia central sobre informa sobre los centros de los datos. Los observables mas comunes incluyen la media y la mediana.
         1.1- Media (también llamada promedio)Es la suma de todos los
              valores dividida por la cantidad de valores. Es afectada
              por los outliers.
         1.2- Mediana: Es el valor que está en el medio cuando los datos
              están ordenados.
         1.3- Moda: Es el valor que más se repite.

      2. La variabilidad informa sobre la dispersión de los datos. Los observables mas comunes incluyen la desviación estándar y el rango intercuartil.
         2.1- La desviación estándar mide cuánto se desvían los valores de
              un conjunto de datos respecto a la media. (chico => cerca de la media)
         2.2- El rango intercuartil es la diferencia entre el tercer
              cuartil (Q3) y el primer cuartil (Q1). Los cuartiles dividen a los datos en 4 partes iguales:
              - Q1 (primer cuartil): el valor que deja el 25% de los datos por debajo de él.
              - Q3 (tercer cuartil): el valor que deja el 75% de los datos por debajo de él.
              Un rango intercuartil pequeño indica que la mayoría de los datos están cerca del centro, mientras que uno grande indica que los datos están más dispersos.

      3. La correlación o variabilidad conjunta le informa sobre la relación entre un par de variables en un conjunto de datos. Las medidas útiles incluyen la covarianza y el coeficiente de correlación.
         3.1- Cov: Si la covarianza es positiva, las dos variables tienden
              a aumentar juntas. (Ejemplo: A medida que sube el ingreso, también lo hace el consumo).Si la covarianza es negativa, una variable tiende a aumentar cuando la otra disminuye. (Ejemplo: A medida que sube la temperatura, la cantidad de ropa de abrigo disminuye).Si la covarianza es cercana a 0, significa que no hay una relación lineal clara entre las dos variables.
         3.2- Varía de -1 a 1; valores cercanos a 1 o -1 indican una
              fuerte relación lineal. El coeficiente de correlación de Pearson mide fuerza y dirección de una relación lineal.
        entonces, La correlación es una versión escalada (estandarizada) de la covarianza, con valores normalizados entre -1 y 1.
        numpy.corrcoef (np.corrcoef()) para calcular el coeficiente de correlación
        np.cov() calcula la covarianza

```python
def media(y):
  suma = 0
  cantidad = 0
  for i in range(len(y)):
    suma+=y[i]
    cantidad+=1
  return suma/cantidad
# definicion desv. estandar = np.std
def desv(y):
  sumatoria = 0
  for i in range(len(y)):
    m = (y[i] - media(y))**2
    sumatoria+=m
  return (sumatoria/len(y))**0.5
```

***DADOS***

* El seed=42 en np.random.seed(seed) sirve para que los números aleatorios generados sean siempre los mismos cada vez que ejecutes el código.

* Sin una semilla, cada vez que corras el programa obtendrás una secuencia diferente de tiradas de dados. Con la semilla fija (42 en este caso, pero puede ser cualquier número), las tiradas serán idénticas en cada ejecución.

```python
# Función para realizar una tirada de dos dados y devolver su suma
def tirar_dados(tiros, seed=42):
    np.random.seed(seed)  # Fijar semilla una sola vez
    random.seed(seed)  # También fijar la semilla para random
    suma_tiros = []
    for _ in range(tiros):
        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
        suma_tiros.append(dado1 + dado2)
    return suma_tiros
# al tirar dos dados la suma de sus valores van del rango de [2,12]. Hago un array con la cantidad de combinaciones posible para que salgan esos valores.
# eje: para que la suma de 2, entinces tengo el (1,1) es la unica opcion, pero para que de 3 tengo (1,2) y (2,1)
# al realizar mas tiros de dos dados independientes , entonces la frecuencia de la suma se aproxima a una distribucion normal
```

Una distribución sesgada positivamente:
La moda (valor que más se repite) suele estar en la zona con mayor densidad de datos → más a la izquierda

La mediana (valor del medio) está más a la derecha que la moda

La media (promedio) es la más afectada por los valores extremos grandes → se "corre" más hacia la derecha

✅ Por eso:
moda < mediana < media

* **Rescalado y normalización**

El rescalado y la normalización de datos implica ajustar los valores de medición de diferentes escalas a una escala y/o distribución común. Como veremos es recomendable realizarlos, especialmente cuando se utilizan como variables de entrada para algoritmos de aprendizaje automático ya que así se asigna el mismo peso a todas las variables, eliminando el impacto de los valores atípicos en los resultados de entrenamiento del modelo.


Existen una gran variedad de métodos, entre los mas comunes están:
- Single feature scaling:Escala los valores a una magnitud unitaria: el vector resultante tiene norma 1.Preserva la dirección del vector, pero no la escala.No garantiza valores entre 0 y 1, ni media 0, ni desviación estándar 1.No es robusto a outliers: si hay un valor muy grande, dominará el denominador.
       = x / ||x||
- Min-max scaling--> Ajustar los datos para que estén dentro de un rango específico, generalmente [0, 1] o [-1, 1]. El Min-max scaling lleva los datos a un rango fijo, sin alterar la forma de la distribución.
       = x - x.min() / x.max() - x.min()
- score normalization -->  Convertir los datos para que tengan una media de 0 y una desviación estándar de 1.
       = (x - x.mean()) / x.std()
- Log scaling: La transformación logarítmica reduce la asimetría de una distribución con sesgo positivo (muchos valores pequeños y pocos muy grandes).Disminuye el peso relativo de los valores extremos sin alterar el orden ni eliminar datos.Es una técnica común para variables como ingresos, precios o poblaciones, donde los outliers son legítimos pero desproporcionados.
       = np.log(x)
- Clipping:Si un valor es menor que el límite inferior, se reemplaza por el límite inferior.Si un valor es mayor que el límite superior, se reemplaza por el límite superior.Los valores dentro del rango se mantienen igual.Mitigar el efecto de outliers extremos, sin eliminarlos completamente del dataset.Evitar que valores atípicos distorsionen estadísticas o el aprendizaje de modelos
       = np.clip(x, x.min(), x.max())

Luego, Rescalado = poner todos los datos en un mismo rango, por ejemplo de 0 a 1.
Normalización = hacer que los datos tengan media 0 y desvío estándar 1 (distribución normal).

* **Resumen Test de hipotesis**

Cuando queremos comparar la efectividad de dos vacunas, como una nueva frente a la ya existente, el análisis estadístico juega un rol fundamental. Uno de los pilares en este tipo de estudios es el **test de hipótesis**, cuyo propósito es determinar si la diferencia observada entre los grupos es lo suficientemente significativa como para atribuirla al efecto del tratamiento y no al azar. En este contexto, formulamos una **hipótesis nula** que plantea que no hay diferencia entre vacunas, y una hipótesis alternativa que sugiere que la nueva vacuna es más efectiva. El test nos ayuda a decidir si podemos rechazar la hipótesis nula con un cierto nivel de confianza.

 Para realizar una comparación justa, necesitamos un **grupo control**, que es aquel que no recibe la nueva vacuna, sino la tradicional o incluso un placebo. Este grupo nos permite establecer una referencia, ayudando a identificar si los efectos en el grupo tratado se deben realmente al nuevo fármaco y no a otros factores como la evolución natural de la enfermedad o incluso efectos psicológicos (como el placebo).

 Es fundamental también que la asignación de los sujetos a los grupos se haga de manera aleatoria. Esta aleatorización reduce el riesgo de sesgos y garantiza que los grupos sean comparables desde el inicio. De este modo, cualquier diferencia observada en los resultados puede atribuirse con mayor confianza al tratamiento, y no a diferencias previas entre los participantes.

 Finalmente, no basta con observar cuál grupo tuvo mejores resultados y decidir en función de eso. Sin herramientas estadísticas, podríamos sacar conclusiones erróneas debido a la variabilidad natural, tamaños de muestra pequeños o sesgos en el diseño. Por eso, el **análisis estadístico** bien estructurado es esencial: nos protege contra errores y asegura que nuestras decisiones se basen en evidencia sólida y no en aparentes coincidencias.

--------------

* **Resumen hipotesis nula**

Dado que tendemos a ver patrones en eventos aleatorios, en los experimentos se necesita evidencia de que una diferencia entre grupos es más extrema de lo que el azar podría explicar. Para eso, partimos de la suposición de que los tratamientos no tienen efectos distintos y que cualquier diferencia observada es producto del azar. Esta suposición inicial se conoce como **hipótesis nula**. El objetivo del análisis es refutar esta hipótesis, demostrando que los resultados de los grupos A y B son más diferentes de lo que cabría esperar bajo el efecto del azar. En otras palabras, **la hipótesis nula suele ser lo opuesto a lo que el investigador o la investigadora busca demostrar.** Por ejemplo, si se desea probar que un nuevo medicamento es más efectivo que uno antiguo, la hipótesis nula establece que no existen diferencias significativas entre ambos.

Una manera de evaluar esto es mediante un **procedimiento de permutación o randomización**: los resultados de los grupos A y B se combinan y se redistribuyen aleatoriamente (en grupos del mismo tamaño original). Se repite este proceso muchas veces para construir una distribución de referencia, llamado **modelo nulo**, que representa la hipótesis de que los grupos son equivalentes e intercambiables y evalúa si los resultados observados pueden explicarse razonablemente bajo ese modelo. En el ejemplo que estamos usando significa que no podemos distinguir entre los tratamientos.

Las pruebas de hipótesis no solo involucran una hipótesis nula, sino también una **hipótesis alternativa**, que representa el efecto que esperamos detectar. Algunos ejemplos:

*   Hipótesis nula: “No hay diferencia entre los promedios de los grupos A y B”

Hipótesis alternativa: “Los promedios de A y B son diferentes”.

*   Hipótesis nula: “A ≤ B”

Hipótesis alternativa: “A > B"

*   Hipótesis nula: “B no es al menos X% mayor que A”

Hipótesis alternativa: “B es al menos X% mayor que A”

* **P-valor**
image.png

el p-valor es una medida estadística que nos indica la probabilidad de obtener un resultado igual o más extremo que el observado, asumiendo que la hipótesis nula es verdadera.

En otras palabras, el p-valor nos dice qué tan compatibles son nuestros datos con la hipótesis nula. Si el p-valor es muy pequeño (por ejemplo, menor a 0.05), significa que es muy poco probable haber obtenido ese resultado solo por azar, y por eso se suele rechazar la hipótesis nula

EJEMPLO: Si estás probando si una vacuna es más efectiva que otra, y obtenés un p-valor de 0.01, eso indica que hay solo un 1% de probabilidad de que esa diferencia se deba al azar. Eso te da evidencia para pensar que la vacuna sí hace una diferencia real.

Uno de los principales problemas con el p-valor es que puede encontrarse una  hipótesis que produzcan una diferencia significativa. Este fenómeno es conocido como $p-hacking$.

Uno de los problemas de fondo es que se suele interpretar el p-valor de manera errónea. Por ejemplo:
*La probabilidad de que el resultado se deba al azar* es una interpretación, un p-valor bajo se asume como una demostración concluyente de un efecto real. Sin embargo, lo que el p-valor realmente representa es:

*la probabilidad de obtener un resultado tan extremo como el observado, asumiendo que el modelo nulo es cierto*

Aunque la diferencia es sutil, es fundamental: un p-valor significativo no proporciona una prueba definitiva de un efecto real, sino solo evidencia contra la hipótesis nula dentro del modelo estadístico utilizado.

### Randomización
Para evaluar la variabilidad aleatoria de una muestra, la randomización de los datos es una técnica estadística en la que se extraen repetidamente muestras de los datos observados. Para eso, se mezclan los datos de todos los grupos, bajo la suposición de que los tratamientos no generan diferencias reales.

Un posible procedimiento de permutació npodria ser:

1. Combinar los resultados de todos los grupos en un solo conjunto de datos.
2. Mezclar aleatoriamente los datos y extraer una muestra del mismo tamaño que el grupo A.
3. Extraer otra muestra, del mismo tamaño que el grupo B, a partir de los datos restantes.
4. Calcular la estadística de interés (por ejemplo, la diferencia entre proporciones) para la nueva distribución muestreada.
5. Repetir los pasos anteriores muchas veces (n iteraciones) para generar la distribución de permutaciones.

Finalmente, se compara la diferencia observada entre grupos (sin permutar) con la distribución de diferencias generada por permutación. Si la diferencia observada está fuera de la mayoría de los valores de la distribución de permutación, se concluye que no es producto del azar. En este caso, la diferencia es **estadísticamente significativa**. En cambio, si la diferencia observada se encuentra dentro del rango esperado por azar, no se puede concluir nada significativo.

Para hacer la randomización (o shuffling) vamos a elegir un **estadístico**. Un estadístico es cualquier medida o valor numérico que se calcula a partir de la muestra de los datos. Se usa para describir y resumir la muestra, y a menudo sirve como una estimación de un parámetro desconocido de la población. Esta variable resume información de los datos observados y nos ayuda a hacer inferencias sobre la población.

* **SSE**

Sum of Squared Errors(Suma de los errores al cuadrado)
Es una medida de qué tan dispersos están los datos respecto a un valor de referencia.Normalmente, el valor de referencia es la media (promedio) de los datos. Me dice cuánta variabilidad hay en los datos respecto al promedio.

Mientras más pequeño es el SSE, más cerca están los datos del promedio ➔ datos más "compactos".Mientras más grande, más dispersos están.

* Calculo:
$$SSE = \sum_{i=1}^{n} (x_i - \bar{x})^2$$
* Incerteza

Si tu problema requiere tener en cuenta que cada dato tiene una incertidumbre propia (por ejemplo, mediciones físicas), entonces el cálculo cambia un poco:
deberías ponderar los errores teniendo en cuenta esas incertidumbres.Sin incertidumbre, todos los errores valen igual.Con incertidumbre, tendrías que pesar más los datos confiables y menos los dudosos.

* SSE ponderado (pesan las incertidumbres)
$$SSE_{\text{ponderado}} = \sum_{i=1}^{n} \left( \frac{x_i - \bar{x}}{\sigma_i} \right)^2$$

* Valor minimo al graficas SSE(valor medio propuesto)
El SSE es mínimo exactamente en el valor donde el promedio propuesto coincide con el promedio real Esa mínima te dice: este es el mejor valor central para minimizar el error cuadrático.
Sea \(\mu\) un valor medio propuesto.

La suma de los errores al cuadrado (SSE) en función de \(\mu\) es:

$$
SSE(\mu) = \sum_{i=1}^{n} (x_i - \mu)^2
$$

El valor de \(\mu\) que minimiza \(SSE(\mu)\) es el promedio de los datos:

$$
\mu = \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$


* **Recta (y=mx+b)**
Cuando tenés una recta (modelo lineal tipo y=mx+b) y datos dispersos, querés saber qué tan bien la recta ajusta a esos datos.El SSE (Suma de los Errores Cuadrados) mide cuánto se alejan los datos de la recta.

Para cada punto $$(𝑥_𝑖,𝑦_𝑖)$$
calculás la distancia vertical entre el dato real 𝑦𝑖 y  el valor predicho por la recta $$𝑦pred =𝑚𝑥𝑖+𝑏$$
Esa diferencia es el error en 𝑥_𝑖:

$$\text{error}_i = y_i - (m x_i + b)$$

Luego, el SSE es la suma de todos esos errores al cuadrado:

$$SSE = \sum_{i=1}^{n} (y_i - (m x_i + b))^2$$

El mejor ajuste es el que minimiza el SSE.Cuando ajustás una recta, por ejemplo con regresión lineal, lo que estás haciendo es buscar la pendiente
𝑚 y el intercepto 𝑏 que minimizan el SSE.
Por qué los errores se elevan al cuadrado:
Para que errores positivos y negativos no se cancelen y para penalizar errores grandes más fuertemente.

`scipy.optimize` es un submódulo de `SciPy` dedicado a la optimización y resolución de ecuaciones. Ofrece una amplia colección de algoritmos para encontrar mínimos o máximos de funciones, resolver sistemas de ecuaciones no lineales, buscar raíces de funciones y ajustar modelos a datos experimentales. Entre sus herramientas se incluyen métodos de optimización local y global, algoritmos de mínimos cuadrados, búsqueda de raíces, programación lineal y optimización con restricciones. Dentro de `optimize`, la función `curve_fit` es especialmente popular para el ajuste de curvas, ya que facilita encontrar automáticamente los parámetros de una función modelo que mejor se adaptan a un conjunto de datos, recibiendo como entrada la función a ajustar, los valores de `x` y las observaciones `y`. Internamente, `curve_fit` utiliza el motor especializado `least_squares`, que trabaja a un nivel más general tomando como entrada una función que devuelve los residuos, es decir, las diferencias entre las predicciones del modelo y los datos observados. Esta diferencia en el enfoque permite que `least_squares` sea más flexible y potente para resolver una amplia variedad de problemas de ajuste no lineal, mientras que `curve_fit` ofrece una interfaz simplificada para casos de uso comunes.

¿Qué nos indica el RMSE?

    Un RMSE bajo significa que las predicciones del modelo son cercanas a los valores reales. Cuanto más bajo sea el RMSE, mejor es el modelo en cuanto a su capacidad para hacer predicciones precisas.

    Un RMSE alto significa que hay una gran discrepancia entre las predicciones y los valores reales. Esto puede indicar que el modelo no está ajustando bien los datos o que está cometiendo errores significativos.

Si el RMSE es el mismo tanto para los buenos como para los malos parámetros iniciales, podría ser una indicación de que el modelo está funcionando bien y el ajuste es robusto. Sin embargo, también es importante verificar cómo se comportan los parámetros ajustados y si el ajuste visualmente se ve bien, especialmente si estás intentando modelar un sistema donde la precisión en los parámetros es crítica.

Considerar otras métricas: Si el RMSE es igual, tal vez se podría mirar otras métricas o técnicas de evaluación, como la gráfica de residuos (donde se analizan las diferencias entre los valores reales y los ajustados) o analizar el comportamiento de los parámetros (si se acercan más a los valores reales en un ajuste versus el otro).

**- ¿Cómo cambia el codigo si usas la función `scipy.optimize.least_squares`?**

(from scipy.optimize import least_squares)


least_squares es una función que resuelve un problema de optimización no lineal minimizando la suma de los cuadrados de los residuos. Es bastante útil para problemas de ajuste de curvas y regresión, ya que busca minimizar la diferencia entre los valores predichos y los datos observados.
Cómo cambia el código con least_squares:

    La función de costo que se pasa a least_squares es la diferencia entre los datos observados y la curva ajustada.

    least_squares no requiere una función explícita para el modelo (como curve_fit), sino que necesitamos definir los residuos de manera explícita.

    Además, least_squares devuelve solo los parámetros ajustados y no los errores estándar, por lo que si se necesitan los errores estándar, se deben calcular por separado.

**- ¿Cómo cambia el codigo si usas la función `scipy.optimize.minimize`?**

(from scipy.optimize import minimize)


La función minimize es más general y flexible. Permite minimizar cualquier función objetivo, no solo la suma de los cuadrados de los errores. Así que podemos usarla para minimizar la suma de los cuadrados de los residuos, pero debemos definir explícitamente esa función de error.
Cómo cambia el código con minimize:

    Definir la función de error: Usamos la misma idea que con least_squares, pero ahora necesitamos proporcionar una función que calcule la suma de los cuadrados de los errores.

    Usar un optimizador: minimize permite elegir varios métodos de optimización. Usaremos el método 'BFGS' (uno de los más comunes y eficientes para este tipo de problemas).

    Ajustar el modelo: Al igual que con least_squares, el resultado de minimize tiene los parámetros ajustados en el atributo .x.

¿Qué opción elegir?

    Si estás trabajando específicamente con un problema de ajuste de curvas y minimización de residuos, least_squares es generalmente la opción más eficiente y directa.

    Si necesitas más control sobre el proceso de optimización o estás trabajando en un problema más general, minimize es más flexible.
